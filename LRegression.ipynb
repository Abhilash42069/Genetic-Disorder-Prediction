```python
import pandas as pd
```

```python
df = pd.read_csv("Gene_Dataset.csv")
```

```python
test_data = df[df["GeneDisorder"].isnull()]
```

```python
test_data
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
y_train = df["GeneDisorder"]
```

```python
y_train
```

```python
X_train = df.drop ("GeneDisorder", axis=1)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
X_test = test_data.drop("GeneDisorder", axis = 1) 
```

```python
X_test
```

# New section

# New section

```python
X_test.shape
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
y_pred.shape
```

```python
test_data.loc[test_data.GeneDisorder.isnull() , "GeneDisorder"] = y_pred
```
